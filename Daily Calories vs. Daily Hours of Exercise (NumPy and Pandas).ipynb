# Daily Calories vs. Daily Hours of Exercise

### Utilize NumPy and Pandas

```python
import random

data = []

for i in range(0,365):
    x = random.randint(500,5000)
    data.append(x)
    
print(data)
```

```python
import numpy as np

calories = np.array(data)
print(calories)
```

```python
calories.shape
```

```python
calories[0]
```

```python
calories.mean()
```

```python
import random

exercise = []

for i in range(0,365):
    x = random.randint(0,12)
    exercise.append(x)

yearly_data = np.array([exercise, calories])

yearly_data
```

```python
yearly_data.shape
```

```python
yearly_data[0][0]
```

```python
import pandas as pd

individuals = pd.DataFrame({'Daily Hours of Exercise':yearly_data[0],
                            'Daily Calories':yearly_data[1]})

individuals
```

```python
individuals.loc[0,'Daily Calories']
```

```python
individuals[individuals['Daily Hours of Exercise'] == 7]
```

```python
exer_average = yearly_data[0].mean()
cals_average = yearly_data[1].mean()

print('Calories AVG: {:.2f}\nExercise AVG: {:.2f}'.format(cals_average, exer_average))
```
