# Microsoft Stocks (Statistical Analysis)

### Utilize SciPy, Matplotlib, Pandas

```python
import pandas as pd

stock = pd.read_csv('/Users/faris/Downloads/MicroSoft.csv',header=None)
stock.rename(columns={0:'Date', 
                      1:'Open',
                      2:'High',
                      3:'Low',
                      4:'Close',
                      5:'Adj Close',
                      6:'Volume'},inplace=True)
stock = stock.drop([0])
stock[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']] = stock[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']].apply(pd.to_numeric)
stock
```

```python
stock.dtypes
```

```python
stock.index[stock['Date']=='2020-01-02'].tolist()
```

```python
stock[8521:]
```

```python
stock_high = stock['High'][8521:]
stock_high
```

```python
minimum_high = stock_high.min()
maximum_high = stock_high.max()
mean_high = stock_high.mean()
median_high = stock_high.median()
mode_high = stock_high.mode()[0]

print('Minimum High: ' + str(minimum_high))
print('Maximum High: ' + str(maximum_high))
print('Mean High: ' + str(mean_high))
print('Median High: ' + str(median_high))
print('Mode High: ' + str(mode_high))
```

```python
from matplotlib import pyplot as plt

fig, gph = plt.subplots(2, 1, figsize = (10,8))

#gph[0].hist(stock_high)
gph[0].set_ylabel('Frequency of High Prices')

gph[0].axvline(x=minimum_high, color = 'gray', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=mean_high, color = 'green', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=median_high, color = 'red', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=mode_high, color = 'orange', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=maximum_high, color = 'gray', linestyle='dashed', linewidth = 5)
  
gph[1].boxplot(stock_high, vert=False)

fig.suptitle('Microsoft Stocks (2020 - Present)')

plt.show()
```

```python
stock_low = stock['Low'][8521:]

minimum_low = stock_low.min()
maximum_low = stock_low.max()
mean_low = stock_low.mean()
median_low = stock_low.median()
mode_low = stock_low.mode()[0]

print('Minimum: ' + str(minimum_low))
print('Maximum: ' + str(maximum_low))
print('Mean: ' + str(mean_low))
print('Median: ' + str(median_low))
print('Mode: ' + str(mode_low))
```

```python
fig, gph = plt.subplots(2, 1, figsize = (10,8))
  
gph[0].hist(stock_low)
gph[0].set_ylabel('Frequency of Low Prices')

gph[0].axvline(x=minimum_low, color = 'gray', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=mean_low, color = 'green', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=median_low, color = 'red', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=mode_low, color = 'orange', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=maximum_low, color = 'gray', linestyle='dashed', linewidth = 5)
  
gph[1].boxplot(stock_low, vert=False)

fig.suptitle('Microsoft Stocks (2020 - Present)')

plt.show()
```

```python
q01 = stock_high.quantile(0.6)
stock_highest = stock_high[stock_high>q01]
stock_highest
```

```python
minimum_highest = stock_highest.min()
maximum_highest = stock_highest.max()
mean_highest = stock_highest.mean()
median_highest = stock_highest.median()
mode_highest = stock_highest.mode()[0]

print('Minimum Highest: ' + str(minimum_highest))
print('Maximum Highest: ' + str(maximum_highest))
print('Mean Highest: ' + str(mean_highest))
print('Median Highest: ' + str(median_highest))
print('Mode Highest: ' + str(mode_highest))
```

```python
fig, gph = plt.subplots(2, 1, figsize = (10,8))
  
gph[0].hist(stock_highest)
gph[0].set_ylabel('Frequency of Highest Prices')

gph[0].axvline(x=minimum_highest, color = 'gray', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=mean_highest, color = 'green', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=median_highest, color = 'red', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=mode_highest, color = 'orange', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=maximum_highest, color = 'gray', linestyle='dashed', linewidth = 5)
  
gph[1].boxplot(stock_highest, vert=False)

fig.suptitle('Microsoft Stocks (2020 - Present)')

plt.show()
```

```python
fig = plt.figure(figsize=(10,4))

stock_highest.plot.density()

plt.title('Microsoft High Stocks 60th+ Quantile (2020 - Present)')

plt.axvline(x = mean_highest, color = 'green', linestyle='dashed', linewidth = 2)
plt.axvline(x = median_highest, color = 'red', linestyle='dashed', linewidth = 2)
plt.axvline(x = mode_highest, color = 'orange', linestyle='dashed', linewidth = 2)

plt.show()
```

```python
for stock_price in [stock_high, stock_low]:
    rang = stock_price.max() - stock_price.min()
    variance = stock_price.var()
    stdeviation = stock_price.std()
    print('')
    print('Range: ' + str(rang))
    print('Variance: ' + str(variance))
    print('Standard Deviation: ' + str(stdeviation))
```

```python
import scipy.stats as stats

density = stats.gaussian_kde(stock_high)
stock_high.plot.density()
```

```python
stdeviation = stock_high.std()
average = stock_high.mean()

density = stats.gaussian_kde(stock_high)
stock_high.plot.density()

x1 = [average-stdeviation, average+stdeviation]
y1 = density(x1)
plt.plot(x1,y1, color='red')
plt.annotate('1 std (68.26%)', (x1[1],y1[1]))

x2 = [average-(stdeviation*2), average+(stdeviation*2)]
y2 = density(x2)
plt.plot(x2,y2, color='orange')
plt.annotate('2 std (95.45%)', (x2[1],y2[1]))

x3 = [average-(stdeviation*3), average+(stdeviation*3)]
y3 = density(x3)
plt.plot(x3,y3, color='maroon')
plt.annotate('3 std (99.73%)', (x3[1],y3[1]))

plt.axvline(average, color='green', linewidth=2)

plt.title('Microsoft High Stocks (2020 - Present)')

plt.show()
```
