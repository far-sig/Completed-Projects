# Coronavirus by Country (Pandas)

### Utilize Pandas

```python
import wget
import pandas as pd

covid = pd.read_csv('/Users/faris/Downloads/COVID-19 Coronavirus.csv.xls',header=None)
covid
```

```python
covid.rename(columns={0:'Country', 
                      1:'Other names',
                      2:'ISO',
                      3:'Population',
                      4:'Continent',
                      5:'Total Cases',
                      6:'Total Deaths',
                      7:'TC1Mpop',
                      8:'TD1Mpop',
                      9:'Death percentage'},inplace=True)
covid
```

```python
covid = covid.drop([0])
covid
```

```python
covid.dtypes
```

```python
covid[['Population', 'Total Cases', 'Total Deaths', 'TC1Mpop', 'TD1Mpop', 'Death percentage']] = covid[['Population','Total Cases', 'Total Deaths', 'TC1Mpop', 'TD1Mpop', 'Death percentage']].apply(pd.to_numeric)
covid.dtypes
```

```python
covid.isnull().sum()
```

```python
covid[covid.isnull().any(axis=1)]
```

```python
covid = covid.dropna(axis=0, how='any')
covid
```

```python
mean_total = covid['Total Cases'].mean()
mean_total
```

```python
covid[covid['Total Cases'] > mean_total]
```

```python
covid[covid['Total Cases'] > mean_total].Population.mean()
```

```python
covidfinal = covid.sort_values('Total Cases', ascending = False)
covidfinal.head()
```
