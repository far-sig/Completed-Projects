# Total Covid Deaths Per 15 Randomly Sampled Countries (Matplotlib)

### Utilize Matplotlib for Statistical Analysis

```python
import wget
import pandas as pd

covid = pd.read_csv('/Users/faris/Downloads/COVID-19 Coronavirus.csv.xls',header=None)
covid.rename(columns={0:'Country', 
                      1:'Other names',
                      2:'ISO',
                      3:'Population',
                      4:'Continent',
                      5:'Total Cases',
                      6:'Total Deaths',
                      7:'TC1Mpop',
                      8:'TD1Mpop',
                      9:'Death percentage'},inplace=True)
covid = covid.drop([0])
covid = covid.dropna(axis=0, how='any')
covid[['Population', 'Total Cases', 'Total Deaths', 'TC1Mpop', 'TD1Mpop', 'Death percentage']] = covid[['Population','Total Cases', 'Total Deaths', 'TC1Mpop', 'TD1Mpop', 'Death percentage']].apply(pd.to_numeric)
covid
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(10,10))

covid2 = covid.sort_values('Total Deaths', ascending = False)
covid2_pop = covid2.sample(n=15)
covid_deaths = covid2_pop['Total Deaths']
plt.hist(covid_deaths)

plt.title('Frequency vs. Distribution of Deaths')
plt.xlabel('Distribution of Deaths')
plt.ylabel('Frequency of Deaths')

plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.boxplot(covid_deaths)

plt.title('Distribution of Deaths')

plt.show()
```

```python
minimum = covid_deaths.min()
maximum = covid_deaths.max()
mean = covid_deaths.mean()
median = covid_deaths.median()
mode = covid_deaths.mode()[0]

print('Minimum: ' + str(minimum))
print('Maximum: ' + str(maximum))
print('Mean: ' + str(mean))
print('Median: ' + str(median))
print('Mode: ' + str(mode))
```

```python
plt.figure(figsize=(10,10))

plt.hist(covid_deaths)

plt.axvline(x=minimum, color = 'gray', linestyle='dashed', linewidth = 5)
plt.axvline(x=mean, color = 'green', linestyle='dashed', linewidth = 5)
plt.axvline(x=median, color = 'red', linestyle='dashed', linewidth = 5)
plt.axvline(x=mode, color = 'orange', linestyle='dashed', linewidth = 5)
plt.axvline(x=maximum, color = 'gray', linestyle='dashed', linewidth = 5)

plt.title('Frequency vs. Distribution of Deaths')
plt.xlabel('Distribution of Deaths')
plt.ylabel('Frequency of Deaths')

plt.show()
```

```python
fig, gph = plt.subplots(2, 1, figsize = (10,8))
  
gph[0].hist(covid_deaths)
gph[0].set_ylabel('Frequency of Deaths')

gph[0].axvline(x=minimum, color = 'gray', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=mean, color = 'green', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=median, color = 'red', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=mode, color = 'orange', linestyle='dashed', linewidth = 5)
gph[0].axvline(x=maximum, color = 'gray', linestyle='dashed', linewidth = 5)
  
gph[1].boxplot(covid_deaths, vert=False)

fig.suptitle('Frequency vs. Distribution of Deaths')

plt.show()
```

```python
fig = plt.figure(figsize=(10,4))

covid_deaths.plot.density()

plt.title('Density of Total Deaths')

plt.axvline(x = mean, color = 'green', linestyle='dashed', linewidth = 2)
plt.axvline(x = median, color = 'red', linestyle='dashed', linewidth = 2)
plt.axvline(x = mode, color = 'orange', linestyle='dashed', linewidth = 2)

plt.show()
```
