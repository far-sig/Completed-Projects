# Total Covid Cases Per Country (Matplotlib)

### Utilize Matplotlib

```python
import wget
import pandas as pd

covid = pd.read_csv('/Users/faris/Downloads/COVID-19 Coronavirus.csv.xls',header=None)
covid.rename(columns={0:'Country', 
                      1:'Other names',
                      2:'ISO',
                      3:'Population',
                      4:'Continent',
                      5:'Total Cases',
                      6:'Total Deaths',
                      7:'TC1Mpop',
                      8:'TD1Mpop',
                      9:'Death percentage'},inplace=True)
covid = covid.drop([0])
covid = covid.dropna(axis=0, how='any')
covid
```

```python
covid[['Population', 'Total Cases', 'Total Deaths', 'TC1Mpop', 'TD1Mpop', 'Death percentage']] = covid[['Population','Total Cases', 'Total Deaths', 'TC1Mpop', 'TD1Mpop', 'Death percentage']].apply(pd.to_numeric)
covid.dtypes
```

```python
#%matplotlib inline

from matplotlib import pyplot as plt

covid2 = covid.sort_values('Total Cases', ascending = False)
covid2_pop = covid2.head(10)

plt.bar(x = covid2_pop['Country'], height = covid2_pop['Total Cases'], color = 'green')

plt.show()
```

```python
plt.bar(x = covid2_pop['Country'], height = covid2_pop['Total Cases'], color = 'red')

plt.title('Total Cases per Country')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.grid(color='black', linestyle='--', linewidth=1, axis='y')
plt.xticks(rotation=90)

plt.show()
```

```python
fig = plt.figure(figsize=(8,8))

plt.bar(x = covid2_pop['Country'], height = covid2_pop['Total Cases'], color = 'blue')

plt.title('Total Cases per Country')
plt.xlabel('Country')
plt.ylabel('Total Cases')
plt.grid(color='black', linestyle='--', linewidth=1, axis='y')
plt.xticks(rotation=90)

plt.show()
```

```python
# Create a figure for 2 subplots (1 row, 2 columns)
fig, gph = plt.subplots(1, 2, figsize = (10,5))

# Create a bar plot of name vs grade on the first axis
covid3_pop = covid2_pop.head()
gph[0].bar(x = covid3_pop['Country'], height = covid3_pop['Total Cases'], color = 'blue')
gph[0].set_title('Total Cases per Country')

# Create a pie chart of pass counts on the second axis
gph[1].pie(covid3_pop['Death percentage'], labels=covid3_pop['Death percentage'])
gph[1].set_title('Death Percentages')
gph[1].legend(covid2_pop['Country'], loc='center', shadow = 'bool')

# Show the figure
plt.show()
```
